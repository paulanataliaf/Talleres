## ¿QUÉ ES?

Este programa es un ejemplo de un autómata celular bidimensional. Este autómata celular particular se llama El juego de la vida.

Un autómata celular es una máquina computacional que realiza acciones basadas en ciertas reglas. Se puede considerar como un tablero que se divide en celdas (como las celdas cuadradas de un tablero de ajedrez). Cada célula puede estar " viva "  o  " muerta " .   Esto se denomina " estado " de la célula. De acuerdo con las reglas especificadas, cada celda estará viva o muerta en el siguiente paso.

## CÓMO FUNCIONA

Las reglas del juego son las siguientes. Cada celda verifica el estado de sí misma y sus ocho vecinos circundantes y luego se establece en vivo o muerto.  Si hay menos de dos vecinos vivos, entonces la célula muere.  Si hay más de tres vecinos vivos, la célula muere.  Si hay 2 vecinos vivos, la celda permanece en el estado en que se encuentra.   Si hay exactamente tres vecinos vivos, la celda se vuelve viva. Esto se hace en paralelo y continúa para siempre.

Hay ciertas formas recurrentes en la vida, por ejemplo, el " planeador "  y el " intermitente " . El planeador se compone de 5 celdas que forman una pequeña forma de punta de flecha, como esta:

`` `texto
   O
    O
  OOO
`` `

Este planeador se moverá por todo el mundo, conservando su forma. Una luz intermitente es un bloque de tres celdas (arriba y abajo o izquierda y derecha) que gira entre las orientaciones horizontal y vertical.
